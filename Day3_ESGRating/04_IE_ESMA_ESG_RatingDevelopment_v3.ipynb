```python
import numpy as np
import pandas as pd

# Number of companies
n_companies = 10000

# Generate company names
company_names = [f'Company_{i}' for i in range(1, n_companies + 1)]

# Generate sectors
sectors = np.random.choice(['Technology', 'Healthcare', 'Finance', 'Energy', 'Consumer Goods'], n_companies)

# Generate countries
countries = np.random.choice(['USA', 'China', 'Germany', 'Japan', 'India'], n_companies)

carbon_emissions = np.random.rand(n_companies) * 100

# Generate ESG-related variables
np.random.seed(0)
data = {
    'company_name': company_names,
    'sector': sectors,
    'country': countries,
    'carbon_emissions': carbon_emissions,
    'energy_efficiency': np.random.rand(n_companies) * 100,
    'water_usage': np.random.rand(n_companies) * 100,
    'waste_reduction': np.random.rand(n_companies) * 100,
    'labor_practices': np.random.rand(n_companies) * 100,
    'community_engagement': np.random.rand(n_companies) * 100,
    'diversity_inclusion': np.random.rand(n_companies) * 100,
    'human_rights': np.random.rand(n_companies) * 100,
    'ethical_sourcing': np.random.rand(n_companies) * 100,
    'supply_chain_transparency': np.random.rand(n_companies) * 100,
    'board_diversity': np.random.rand(n_companies) * 100,
    'environmental_management': np.random.rand(n_companies) * 100,
    'green_buildings': np.random.rand(n_companies) * 100,
    'renewable_energy': np.random.rand(n_companies) * 100,
    'executive_pay': np.random.rand(n_companies) * 100,
    'political_contributions': np.random.rand(n_companies) * 100,
    'tax_transparency': np.random.rand(n_companies) * 100,
    'data_privacy': np.random.rand(n_companies) * 100,
    'product_safety': np.random.rand(n_companies) * 100,
    'cyber_security': np.random.rand(n_companies) * 100
}

# Create DataFrame
df = pd.DataFrame(data)

```

```python
import numpy as np

esg_factors_direction = {
    'carbon_emissions': -1,
    'energy_efficiency': 1,
    'water_usage': -1,
    'waste_reduction': 1,
    'labor_practices': 1,
    'community_engagement': 1,
    'diversity_inclusion': 1,
    'human_rights': 1,
    'ethical_sourcing': 1,
    'supply_chain_transparency': 1,
    'board_diversity': 1,
    'environmental_management': 1,
    'green_buildings': 1,
    'renewable_energy': 1,
    'executive_pay': -1,
    'political_contributions': -1,
    'tax_transparency': 1,
    'data_privacy': 1,
    'product_safety': 1,
    'cyber_security': 1
}

print(esg_factors_direction)
```


# Applying transformation to normalize variables by sector
* 0 corresponds to the 1.6% worst
* 1 corresponds to the next 3.2% better than 0, but worst than all the rest
* ...
* 9 corresponds to the 3.2% almost best
* 10 corresponds to the 1.6% best
* quantiles = [0.016, 0.048, 0.113, 0.242, 0.5, 0.758, 0.887, 0.952, 0.984, 1]


```python
quantiles = [0.016, 0.048, 0.113, 0.242, 0.5, 0.758, 0.887, 0.952, 0.984, 1]
labels = range(1, len(quantiles)) # You should have 9 bin edges (quantiles) and 8 labels

for col in data.keys():
    if col not in ['company_name', 'sector', 'country']:
        df[f'{col}_group'] = df.groupby('sector')[col].transform(
            lambda x: pd.qcut(x*esg_factors_direction[col], q=quantiles, labels=labels, duplicates='drop')).fillna(9)


print(df.head())
```

# Model calculation using Principal Component Analysis - Unsupervised

This script performs the following steps:

Selects only the columns ending with _group from the DataFrame df.

Standardizes the data using StandardScaler.

Applies PCA to the standardized data.

Creates a DataFrame with the principal components.

Combines the original DataFrame with the PCA components for further analysis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming 'df' is the DataFrame containing your data

# Select only the columns ending with '_group'
group_columns = [col for col in df.columns if col.endswith('_group')]

# Extract the data for PCA
X = df[group_columns]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# Calculate explained variance
explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Elbow Curve for PCA')
plt.grid(True)
plt.show()
```

```python
# Assuming PCA has been performed as shown in the previous code snippet

# Print the first principal component
print(f'First principal component:\n{pca.components_[0]}')
```

```python
# Create a DataFrame for the principal component loadings
loadings_df = pd.DataFrame(pca.components_, columns=group_columns)

# Print the loadings of the first principal component sorted by absolute value
print('First principal component loadings sorted by absolute value:')
print(loadings_df.iloc[0].abs().sort_values(ascending=False))
```
